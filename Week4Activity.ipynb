```python
#Exercise 2 
import cv2
import numpy as np

def crop_grid(img, num_horizontal_grid, num_vertical_grid, line_color):
    # Copy the source image
    img_copy = img.copy()

    # Get the dimensions of the image
    height, width = img.shape[:2]

    # Calculate the patch size
    patch_height = height // num_vertical_grid
    patch_width = width // num_horizontal_grid

    # Draw grid lines
    for i in range(1, num_vertical_grid):
        cv2.line(img_copy, (0, i * patch_height), (width, i * patch_height), line_color, 1)
    for j in range(1, num_horizontal_grid):
        cv2.line(img_copy, (j * patch_width, 0), (j * patch_width, height), line_color, 1)

    return img_copy

# Load the image
img = cv2.imread('D:\MiniProjectImages\dog.jfif')

# Divides the image into a 3x2 grid with red grid lines
cropped_img = crop_grid(img, 3, 2, (0, 0, 255))

# Display the patched image
cv2.imshow('Patched Image', cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Exercise 3 
import cv2

# Load the base images
img1 = cv2.imread("D:\MiniProjectImages\lena.jfif")
img2 = cv2.imread("D:\MiniProjectImages\coins.jfif")

# Define different alpha values
alpha_values = [0.1, 0.3, 0.5, 0.7, 0.9]

# Resize the second image to match the dimensions of the first image
img2_resized = cv2.resize(img2, img1.shape[:2][::-1])

# Display the image sequences with different alpha values
for alpha in alpha_values:
    blended_image = cv2.addWeighted(img1, alpha, img2_resized, 1-alpha, 0)
    cv2.imshow(f"Alpha: {alpha:.1f}", blended_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
