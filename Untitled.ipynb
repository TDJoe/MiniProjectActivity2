```python
import cv2

# Load the video
video_path = 'D:/MiniProjectVideo/img_pexels.mp4'
cap = cv2.VideoCapture(video_path)

# Get the video's frame dimensions and frame rate
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_rate = 30  # Arbitrary frame rate, you can adjust as needed

# Define the new dimensions for smaller frames
new_width = 320
new_height = 240

# Create a VideoWriter object to save the resized video
output_path = 'smaller_img_pexels.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_path, fourcc, frame_rate, (new_width, new_height))

# Read and resize each frame, then display and save
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # Resize the frame
    resized_frame = cv2.resize(frame, (new_width, new_height))

    # Display the resized frame
    cv2.imshow('Smaller Frame', resized_frame)
    
    # Save the resized frame
    out.write(resized_frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and writer objects
cap.release()
out.release()

# Destroy all windows
cv2.destroyAllWindows()
```

```python
import cv2

# Load the original image
image_path = "D:\MiniProjectImages\dog.jfif"
image = cv2.imread(image_path)

# Define the scale factor for enlargement
scale_factor = 2

# Upscale using linear interpolation
linear_scaled_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_LINEAR)

# Upscale using cubic interpolation
cubic_scaled_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_CUBIC)

# Upscale using nearest neighbor interpolation
nn_scaled_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_NEAREST)

# Display the upscaled images
cv2.imshow("Linear Interpolation", linear_scaled_image)
cv2.imshow("Cubic Interpolation", cubic_scaled_image)
cv2.imshow("Nearest Neighbor Interpolation", nn_scaled_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
