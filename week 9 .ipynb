```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the image
image = cv2.imread('zebra.jfif')

# Convert the image to different color spaces
image_bgr = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Create a list of color space images
images = [image_bgr, image_hsv, image_lab]
color_spaces = ['BGR', 'HSV', 'LAB']

# Define the value of k you want to test
k_value = 2
# Define a function to perform k-means clustering and display the results
def perform_kmeans(image, color_space, k):
    # Reshape the image into a 2D array of pixels
    pixels = image.reshape(-1, 3)

    # Apply k-means clustering
    kmeans = KMeans(n_clusters=k, random_state=0).fit(pixels)

    # Reshape the labels back to the original image dimensions
    segmented_image = kmeans.labels_.reshape(image.shape[:2])

    # Display the segmented image
    plt.figure(figsize=(8, 4))
    plt.imshow(segmented_image, cmap='jet')
    plt.title(f'Segmentation (K={k}) - {color_space}')
    plt.axis('off')
    plt.show()

# Perform k-means clustering and display the results for the specified k value
for color_space_index, image in enumerate(images):
    perform_kmeans(image, color_spaces[color_space_index], k_value)
```
